```python
import pandas as pd
pd.options.display.max_colwidth = 300
```

```python
# load in data
df = pd.read_csv('Data/String_Cleaning.csv', header = None)
df.head()
```

# Phone number
- In our data, all phone numbers include an area code,
- It may or may not have () around the area code
- After 3, 3, and 4 digits, may be separated by - or .
- **Hint: the ? means the preceding pattern is optional (may or may not be present)**

```python
# Extract phone number
# add pattern for phone numbers between ()
phone = '()'
df['phone_number']= df[1].str.extract(phone)
df
```

# State
- All state values in our data are 2 capital letters

```python
# Extract State
# add pattern for state between ()
state = '()'
df['state']= df[1].str.extract(state)
df
```

# City
- all start with a capital letter
- May have a space space
- *Hint \s is a white space*

```python
# City names
# add pattern for city between ()
city = '()'
df['city'] = df[1].str.extract(city)
df
```

# URL
- all start with http or https
- followed by ://
- followed by any number of characters (no spaces)
- **Hint: \S is any non-space character**

```python
# URL
# add pattern for url between ()
url = '()'
df['url'] = df[1].str.extract(url)
df
```
