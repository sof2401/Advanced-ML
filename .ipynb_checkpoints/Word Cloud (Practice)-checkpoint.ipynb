<h2><span style="background-color: initial; font-family: Gotham-Rounded-Bold; font-size: 24px;">Word Cloud (Practice)</span></h2><p>Create a word cloud from the Bram Stoker's Dracula text found <a href="https://drive.google.com/file/d/1qfGcziOsz6Wr069XS8lGr_jpt0FIZwPF/view?usp=sharing" target="_blank">here.</a> Refer the top of the lesson "Word Cloud" for assistance in reading in the .txt. data.</p><p>The word cloud should:</p><ul><li>Include standard stopwords and the following custom stopwords: 'said', 'will', 'one'</li><li>Use the "coolwarm" colormap</li><li>Only include words that are at least 3 characters long</li><li>Have dimensions of 800 X 400 in a figure size of (10, 5)</li><li>Include the title "Dracula"</li></ul><p><br></p><p>Solution</p><p><br></p>

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Define the filepath
fpath= "Data/Dracula.txt"
# Use with open syntax--avoid error on Windows
with open(fpath, encoding = "utf-8") as f:
    txt = f.read()
# Report length of the text
print(f"There are {len(txt)} characters in the full text.\n")
```

```python
# Preview first 1,000 characters (printing)
print(txt[:1000])
```

## Default Cloud

```python
# Create an instance of a WordCloud and use the generate method
cloud = WordCloud(random_state = 123).generate(txt)
plt.imshow(cloud);
plt.axis('off');
```

## Include custom stopwords

```python
# Check list of stopwords
from wordcloud import STOPWORDS
```

```python
# Add custom stopwords
stops = [*STOPWORDS, 'said', 'will', 'one']
```

```python
# Remove stopwords from the wordcloud
cloud = WordCloud(random_state=123, stopwords=stops).generate(txt)
plt.imshow(cloud);
plt.axis('off');
```

## Change Colormapping

```python
# add colormap
cloud = WordCloud(random_state=123, 
                  stopwords=stops, 
                  colormap = 'coolwarm'
).generate(txt)
plt.imshow(cloud);
plt.axis('off');
```

## Set minimum word length

```python
# Limit to words of 3 characters or more
# add colormap
cloud = WordCloud(random_state=123, 
                  stopwords=stops, 
                  colormap = 'coolwarm',
                  min_word_length = 3
).generate(txt)
plt.imshow(cloud);
plt.axis('off');
```

## Change size

```python
# Limit to words of 3 characters or more
# add colormap
cloud = WordCloud(random_state=123, 
                  stopwords=stops, 
                  colormap = 'coolwarm',
                  min_word_length = 3,
                  width = 800,
                  height = 400
).generate(txt)
plt.figure(figsize = (10, 5))
plt.imshow(cloud);
plt.axis('off');
```

```python
# Add a title
cloud = WordCloud(random_state=123, 
                  stopwords=stops, 
                  colormap = 'coolwarm',
                  min_word_length = 3,
                  width = 800,
                  height = 400
).generate(txt)
plt.figure(figsize = (10, 5))
plt.title('Dracula')
plt.imshow(cloud);
plt.axis('off');
```
