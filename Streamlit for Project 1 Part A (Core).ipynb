```python
# Import standard packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',100)
```

```python
# Load data 
fpath =  "Data/sales_predictions_2023.csv"
df = pd.read_csv(fpath)
df.head()
```

```python
df['Item_Fat_Content'].unique()
```

```python
df['Item_Fat_Content']= df['Item_Fat_Content'].replace('low fat', 'Low Fat')
```

```python
df['Item_Fat_Content']= df['Item_Fat_Content'].replace('LF', 'Low Fat')
```

```python
df['Item_Fat_Content']= df['Item_Fat_Content'].replace('reg', 'Regular')
```

```python
df= df.drop(columns=['Item_Identifier', 'Outlet_Identifier', 'Outlet_Establishment_Year'])
df.head()
```

```python
df.to_csv('sales_prediction_2023_modified.csv', index=False)
```

```python
# Histogram
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Item_Type')
plt.xticks(rotation=90)
```

```python
# Function for returning a histogram fig
def plot_hist(df, column):
    fig, ax = plt.subplots(figsize=(12, 6))
    sns.histplot(data=df, x=column, ax=ax)
    plt.xticks(rotation=90)
    return fig
```

```python
fig = plot_hist(df, 'Item_Type')
```
