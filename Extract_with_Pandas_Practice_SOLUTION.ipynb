```python
import pandas as pd
url = 'https://www.fatsecret.com/calories-nutrition/food/snacks'
tables = pd.read_html(url)
len(tables)
```

```python
# Save the tables[3] object as a df
df = tables[3]
df.head()
```

```python
# Remove the unwanted rows (index 0,12,27)
df = df.drop(index = [0,12,27])
df
```

```python
df['Prot(g)'].value_counts()
```

```python
# Replace the - with a 0
df['Prot(g)'] = df['Prot(g)'].replace('-', 0)
df['Prot(g)'].value_counts()
```

```python
df.info()
```

```python
# Convert nutrition columns to float
convert = {'Fat(g)': float, 'Carbs(g)': float, 'Prot(g)': float, 'Calories': float}
df = df.astype(convert)
```

```python
df.describe().round(2)
```
